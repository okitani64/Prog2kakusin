```python
import numpy as np
from matplotlib import pyplot as plt
import math

print('y=Asin(x+C/Bπ)のグラフを表示します')
print('A(係数),B(分母),C(分子)に当てはまる数値を入力してください')

A=int(input('Aの値は？'))
B=int(input('B(分母)の値は？'))
C=int(input('C(分子)の値は？'))

x = np.linspace(-12,12,1000)
y = A * np.sin( x + C*np.pi/B )

plt.plot(x , y)
plt.grid()
plt.show()
```
