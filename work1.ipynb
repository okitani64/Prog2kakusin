```python
# pandasのインポート
import pandas as pd

#Excelで作ったcsvファイルの読み込み
from google.colab import files
uploaded=files.upload()

#csvファイルの読み込み
df = pd.read_csv('mogitest.csv')
df
```

```python
#点数を聞く
study=[ '国語' , '数学' , '化学' , '物理' , '地理' , '英語']
scores=[]
total_point=0
for curriculum in study :
  score=int(input(f'{curriculum}の点数は何点ですか？'))
  scores.append(score)
  total_point+=score

#csvファイルにその人の点数の情報を入れる(行の追加として)
new_row = { '国語':scores[0] , '数学':scores[1] , '化学':scores[2] , '物理':scores[3] , '地理':scores[4] , '英語':scores[5] }
df_new = pd.DataFrame ( new_row , index=[0] )
df = pd.concat ( [df , df_new] , ignore_index=True )
df
```

```python
#合計の平均点を求める
df['合計'] = df['国語'] + df['数学'] + df['化学'] + df['物理'] + df['地理'] + df['英語']

total=df['合計'].sum()/len(df['合計'])
print( f'合計の平均点は{total}点です')

#各教科の平均点を求める
for subject in study :
  average = df[subject].sum() / len(df[subject])
  print(f'{subject}の平均点は{average}点です')
```

```python
#合計の最高点・最低点を求める
top=df['合計'].max()
worst=df['合計'].min()
print(f'合計の最高点は{top}点,最低点は{worst}点です')

#各教科の最高点・最低点を求める
for subject in study :
  max=df[subject].max()
  min=df[subject].min()
  print(f'{subject}の最高点は{max}点,最低点は{min}点です')
```

```python
#順位を求める関数
def rank (sort,subject,mypoint) :
  rank=0
  for i in sort[subject] :
    if i > mypoint :
      rank+=1
  return rank+1

#合計の順位を求める
df_total = df.sort_values('合計' , ascending=False)
total=rank(df_total,'合計',total_point)
print(f'あなたの合計の順位は{total}位です')

#各教科の順位を求める
i=0
for subject in study :
  df=df.sort_values(subject,ascending=False)
  total=rank(df,subject,scores[i])
  print(f'あなたの{subject}の順位は{total}位です')
  i+=1
```
